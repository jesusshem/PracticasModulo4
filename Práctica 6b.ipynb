# Librerías

```r
packages <- c("stats", "cluster", "factoextra")
lapply(packages, library, character.only = T)

library(NbClust)
```

# Importación de datos

```r
df <- read.csv('num_mesnajes.csv')
```

```r
head(df, 6)
```

# Clustering

```r
kmeans(df, 5, iter.max = 30, nstart = 1)
```

## Determinando el número óptimo de clústers

### Elbow method

```r
fviz_nbclust(df, kmeans, method = "wss") + geom_vline(xintercept = 3, linetype = 2) + labs(subtitle = "Elbow method")
```

### Silhouette method

```r
fviz_nbclust(df, kmeans, method = "silhouette") + labs(subtitle = "Silhouette method")
```

### La "regla de la mayoría"

```r
nb <- NbClust(df, distance = "euclidean", min.nc = 2, max.nc = 10, method = "kmeans")
```

Visualizar el resultado de el cálculo de la mayoría de los estadísticos:

```r
fviz_nbclust(nb)
```
